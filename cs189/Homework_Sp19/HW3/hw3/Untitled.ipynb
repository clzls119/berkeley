```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
from scipy.stats import multivariate_normal
```

## 2. Isocontours of Normal Distributions

### a

```python
x = np.arange(-4.0, 4.0, 0.1)
y = np.arange(-4.0, 4.0, 0.1)
X, Y = np.meshgrid(x, y)
```

```python
mu, var = np.array([1,1]), np.array([[1,0], [0,2]])
pos = np.dstack((X, Y))
rv = multivariate_normal(mu, var)

plt.figure(figsize=(8,6))
CS = plt.contourf(X, Y, rv.pdf(pos));
```

### b

```python
mu, var = np.array([-1,2]), np.array([[2,1], [1,3]])
pos = np.dstack((X, Y))
rv = multivariate_normal(mu, var)

plt.figure(figsize=(8,6))
CS = plt.contourf(X, Y, rv.pdf(pos))
plt.xlim(-4,3)
plt.ylim(-2);
```

### c

```python
mu1, var1 = np.array([0,2]), np.array([[2,1], [1,1]])
mu2, var2 = np.array([2,0]), np.array([[2,1], [1,1]])

pos = np.dstack((X, Y))
rv = multivariate_normal(mu1 - mu2, var1 + var2)

plt.figure(figsize=(8,6))
CS = plt.contourf(X, Y, rv.pdf(pos))
plt.ylim(-1.5);
```

### d

```python
mu1, var1 = np.array([0,2]), np.array([[2,1], [1,1]])
mu2, var2 = np.array([2,0]), np.array([[2,1], [1,3]])

pos = np.dstack((X, Y))
rv = multivariate_normal(mu1 - mu2, var1 + var2)

plt.figure(figsize=(8,6))
CS = plt.contourf(X, Y, rv.pdf(pos))
plt.ylim(-2.5);
```

### e

```python
mu1, var1 = np.array([1,1]), np.array([[2,0], [0,1]])
mu2, var2 = np.array([-1,-1]), np.array([[2,1], [1,2]])

pos = np.dstack((X, Y))
rv = multivariate_normal(mu1 - mu2, var1 + var2)

plt.figure(figsize=(8,6))
CS = plt.contourf(X, Y, rv.pdf(pos))
plt.ylim(-1.5);
```

## 3. Eigenvectors of the Gaussian Covariance Matrix

Seed = 123

```python
import seaborn as sns
```

```python
N = 100

np.random.seed(123)

X1 = np.random.normal(3, 9, N)
X2 = np.random.normal(4, 4, N) + X1 * 0.5
```

### a. Mean of Sample

```python
Z = np.array([X1, X2])

Z.mean(axis=1)
```

### b. Covariance

```python
cov = np.cov(Z)
cov
```

### c. Eigenvalues and Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(cov)

eigenvalues
```

```python
eigenvectors
```

### d. Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(X1, X2)

proj = np.dot(Z.T, eigenvectors)

sigma = proj.std(axis=0).mean()

for ev in eigenvectors:
    start, end = proj.mean(axis=0), proj.mean(axis=0) + sigma * ev
    plt.annotate(
        '', xy=end, xycoords='data',
        xytext=start, textcoords='data',
        arrowprops=dict(facecolor='red', width=2.0))
    
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('X1')
plt.ylabel('X2');
```

### e

```python
rot = np.dot(eigenvectors, Z - Z.mean(axis=1, keepdims=True))

plt.figure(figsize=(8,6))
plt.scatter(rot[0], rot[1])

plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('X1')
plt.ylabel('X2');
```
