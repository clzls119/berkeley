## Isocontours of Normal Distributions

```python
x = np.arange(-4.0, 4.0, 0.1)
y = np.arange(-4.0, 4.0, 0.1)
X, Y = np.meshgrid(x, y)
```

## Eigenvectors of the Gaussian Covariance Matrix

Seed = 123

```python
import seaborn as sns
```

```python
N = 100

np.random.seed(123)

X1 = np.random.normal(3, 9, N)
X2 = np.random.normal(4, 4, N) + X1 * 0.5
```

### a. Mean of Sample

```python
Z = np.array([X1, X2])

Z.mean(axis=1)
```

### b. Covariance

```python
cov = np.cov(Z)
cov
```

### c. Eigenvalues and Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(cov)

eigenvalues
```

```python
eigenvectors
```

### d. Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(X1, X2)

proj = np.dot(Z.T, eigenvectors)

sigma = proj.std(axis=0).mean()

for ev in eigenvectors:
    start, end = proj.mean(axis=0), proj.mean(axis=0) + sigma * ev
    plt.annotate(
        '', xy=end, xycoords='data',
        xytext=start, textcoords='data',
        arrowprops=dict(facecolor='red', width=2.0))
    
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('X1')
plt.ylabel('X2');
```

### e

```python
rot = np.dot(eigenvectors, Z - Z.mean(axis=1, keepdims=True))

plt.figure(figsize=(8,6))
plt.scatter(rot[0], rot[1])

plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.xlabel('X1')
plt.ylabel('X2');
```

## Gaussian Classifier

```python
def split(data, size):
    from math import floor
    
    dat = data.copy()
    
    # for reproducibility
    np.random.seed(24)
    
    # shuffle copied data
    np.random.shuffle(dat)
    
    if type(size) == float:
        size = floor(len(dat) * size)

    # training_data, validation_data, training_label, validation_label
    return dat[size:, :-1], dat[:size, :-1], dat[size:, -1], dat[:size, -1]
```

```python
# p(x | c)
def lda(x, mean, var, prior, d, cls):
    
    probs = np.array([])
    
    for i in range(x.shape[0]):
        prob = []
        for j in range(cls):

            z = x[i] - mean[j]

            y = np.linalg.lstsq(var, z)[0]

            prob.append((z @ y) * -.5 + np.log(prior[j]) - d * np.linalg.det(cov))

        probs = np.append(probs, np.argmax(prob))
        
    return probs
```

```python
# p(x | c)
def qda(x, mean, var, prior, d, cls):
    
    probs = np.array([])
    
    for i in range(x.shape[0]):
        prob = []
        for j in range(cls):

            z = x[i] - mean[j]

            y = np.linalg.lstsq(var, z)[0]

            prob.append((z @ y) * -.5 + np.log(prior[j]) - d * np.linalg.det(cov))

        probs = np.append(probs, np.argmax(prob))
        
    return probs
```

```python
mnist_data = loadmat('./mnist-data/mnist_data.mat')

training_data = mnist_data['training_data'] / 255
training_label = mnist_data['training_labels']
testing_data = mnist_data['test_data'] / 255

data = np.hstack((training_data, training_label))

X_train, X_val, y_train, y_val = split(data, 10000)

N = len(training_label)
dim = training_data.shape[1]
cls = len(np.unique(training_label))
```

```python
data_prior = {i: (training_label==i).sum() / N for i in range(cls)}
data_mean = {i: X_train[np.where(y_train == i)].mean(axis=0) for i in range(cls)}
data_cov = {i: np.cov(X_train[np.where(y_train == i)].T) for i in range(cls)}

cov = np.zeros((dim, dim))

for k, v in data_cov.items():
    cov += v
    
cov /= cls
```

### SPAM-HAM QDA

```python
spam = loadmat('./spam-data/spam_data.mat')

spam_training_data = spam['training_data']
spam_training_label = spam['training_labels']
spam_test = spam['test_data']
spam_training = np.hstack((spam_training_data, spam_training_label))
```

```python
X_train, X_val, y_train, y_val = split(spam_training, 0.2)
```

```python
X_train.shape
```

```python
N = len(spam_training_label)
dim = X_train.shape[1]
cls = len(np.unique(spam_training_label))
```

```python
data_prior = {i: (spam_training_label==i).sum() / N for i in range(cls)}
data_mean = {i: X_train[np.where(y_train == i)].mean(axis=0) for i in range(cls)}
data_cov = {i: np.cov(X_train[np.where(y_train == i)].T) for i in range(cls)}

cov = np.zeros((dim, dim))

for k, v in data_cov.items():
    cov += v
    
cov /= cls
```

```python
probs = qda(X_val, data_mean, cov, data_prior, dim, cls)
```

```python
# check for validation accuracy
sum(probs == y_val) / len(probs)
```

```python
pred = qda(spam_test, data_mean, cov, data_prior, dim, cls)
```

```python
sub = pd.DataFrame(data={'Id':np.arange(1, len(spam_test)+1), 'Category':pred})
sub['Category'].astype = 'int'
sub.to_csv('./spam_submission.csv', index=False)
```
