```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline
import sklearn.linear_model
from sklearn.model_selection import train_test_split
```

```python
######## YOUR CODE FOR (c) ##########

## Input: original dimension d,
## Input: embedding dimension k
## Output: d x k random
## Gaussian matrix J with entry-wise
## variances 1/k so that, 
## for any row vector z^T in R^d, 
## z^T J  is a random features embedding for z^T
def random_JL_matrix(d, k):
    return np.random.normal(loc=0, scale=np.sqrt(1/k), size=(d,k))


## Input: n x d data matrix X
## Input: embedding dimension k
## Output: d x k matrix V
## with orthonormal columns 
## corresponding to the top k right signular vectors 
## of X. Thus, for a row vector z^T in R^d
## z^T V  is the projection of z^T 
## onto the the top k right-singular vectors of X,
def pca_embedding_matrix(X, k):
    u, s, v = np.linalg.svd(X, 0)

    return v.T[:k].T

#     v = v.T[:k].T
#     s = s * np.eye(len(s))
#     for i in range(k,len(s)):
#         s[i][i] = 0
#     return u @ s @ v
    
######## END YOUR CODE FOR (c) ##########
```

```python
def rand_embed_accuracy_split(X, y, k):
    '''
    Fitting a k dimensional feature set obtained
    from random embedding of X, versus y
    for binary classification for y in {-1, 1}
    '''
    
    # test train split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
    
    # random embedding
    _, d = X.shape
    J = random_JL_matrix(d,k)
    rand_embed_X = linear_feature_transform(X_train, J)
    
    # fit a linear model
    line = sklearn.linear_model.LinearRegression(fit_intercept=False)
    line.fit(rand_embed_X, y_train)
    
    # predict y
    y_pred=line.predict(linear_feature_transform(X_test, J))
    
    # return the test error
    return 1-np.mean(np.sign(y_pred)!= y_test)

def pca_embed_accuracy_split(X, y, k):
    '''
    Fitting a k dimensional feature set obtained
    from PCA embedding of X, versus y
    for binary classification for y in {-1, 1}
    '''

    # test-train split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

    # pca embedding
    V = pca_embedding_matrix(X_train, k)
    pca_embed_X = linear_feature_transform(X_train,V)
                
    # fit a linear model
    line = sklearn.linear_model.LinearRegression(fit_intercept=False)
    line.fit(pca_embed_X, y_train)
    
     # predict y
    y_pred=line.predict(linear_feature_transform(X_test,V))
    

    # return the test error
    return 1-np.mean(np.sign(y_pred)!= y_test)

# applies the linear transformation N 
# to the rows of X, via X.dot(N)
def linear_feature_transform(X, N):
    return X.dot(N)

# uses pca_embedding_matrix method
# to embed the rows of X onto the first
# k principle components
def pca_embedding(X,k):
    P = pca_embedding_matrix(X, k)
    return linear_feature_transform(X,P)

# uses random_JL_matrix method
# to transform rows of X
# by a k-dimensional JL transform
def random_embeddings(X,k):
    P = random_JL_matrix(X.shape[1], k)
    return linear_feature_transform(X,P)
```

```python
######## LOADING THE DATASETS #########

# to load the data:
data = np.load('data1.npz')
data2 = np.load('data2.npz')
data3 = np.load('data3.npz')

X = data['X']
y = data['y']
X2 = data2['X']
y2 = data2['y']
X3 = data3['X']
y3 = data3['y']
dat = [X, X2, X3]
vals = [y, y2, y3]
n, d = X.shape
n_trials = 10
```

```python
pdat1 = pca_embedding(X, 2)
pdat2 = pca_embedding(X2, 2)
pdat3 = pca_embedding(X3, 2)

rdat1 = random_embeddings(X, 2)
rdat2 = random_embeddings(X2, 2)
rdat3 = random_embeddings(X3, 2)

plts = [[pdat1, pdat2, pdat3], [rdat1, rdat2, rdat3]]
```

```python
pca_errors, random_errors = [], []
for i in range(3):
    perr = []
    rerr = []
    for k in range(1, d+1):
        perr.append(pca_embed_accuracy_split(dat[i], vals[i], k))
        r = 0
        for t in range(n_trials):
            r += rand_embed_accuracy_split(dat[i], vals[i], k)
        rerr.append(r/10)
    pca_errors.append(perr)
    random_errors.append(rerr)
```

```python
x = np.arange(1, d+1, 1)

plt.figure(figsize=(12,8))
ind = 1
for i in range(2):
    for j in range(3):
        ax = plt.subplot(2, 3, ind)
        ind += 1
        if i == 0:
            plt.plot(x, pca_errors[j])
            ax.set_ylabel('Accuracy')
            ax.set_xlabel('k-PCA FEATURES ON DATA {}'.format(j+1))
        else:
            plt.plot(x, random_errors[j])
            ax.set_ylabel('Accuracy')
            ax.set_xlabel('k-DIM RANDOM ACCURACY ON DATA {}'.format(j+1))
        
plt.tight_layout()
plt.show();
```

```python
ind = 1
plt.figure(figsize=(12,8))
for i in range(2):
    for j in range(3):
        ax = plt.subplot(2, 3, ind)
        ind += 1
        plt.scatter(plts[i][j].T[0], plts[i][j].T[1])
        
        if i == 0:
            ax.set_xlabel('PCA_Embedding data {}'.format(j+1))
        else:
            ax.set_xlabel('RANDOM_Embedding data {}'.format(j+1))

plt.tight_layout()
plt.show();
```

```python
pmat_singular = []
rmat_singular = []

for i in range(3):
#     for k in range(1, d+1):
    pmat = pca_embedding(dat[i], d)
    rmat = random_embeddings(dat[i], d)

    _, ps, _ = np.linalg.svd(pmat)
    _, rs, _ = np.linalg.svd(rmat)

    pmat_singular.append(ps)
    rmat_singular.append(rs)
```
